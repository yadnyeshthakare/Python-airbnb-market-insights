```python
##Heat map
ABC=df[ ['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
```

```python
ABC
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data=ABC,annot=True)     #annot=True: Shows numbers inside the squares
```

```python
# Correlation Value	


# +1 to +0.5 :Strong positive relation =we can confidently increase rent if the related factor (like availability or reviews) also goes up

# +0.5 to 0 :Weak positive relation	=Slight influence — increasing rent might work, but not a strong signal

# 0	      :No relation	            = That factor doesn’t affect price much

# -0.5 to -1 :Negative relation       =if this factor goes up, price goes down — do not increase rent blindly
```
